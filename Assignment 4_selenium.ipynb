```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

df = pd.DataFrame(columns=['Rank', 'Name', 'Artist', 'Date', 'Views'])

try:
    ranks = driver.find_elements(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody/tr/td[1]')
    names = driver.find_elements(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody/tr/td[2]')
    artists = driver.find_elements(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody/tr/td[3]')
    dates = driver.find_elements(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody/tr/td[5]')
    views = driver.find_elements(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody/tr/td[4]')
    
    df["Rank"] = [rank.text for rank in ranks]
    df["Name"] = [name.text for name in names]
    df["Artist"] = [artist.text for artist in artists]
    df["Date"] = [date.text for date in dates]
    df["Views"] = [view.text for view in views]
    display(df)
    
except NoSuchElementException:
    print("The Elemnt was Not Found!!!")

```

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://github.com/")
driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/div/nav/ul/li[3]/button").click()
driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a").click()

repo_titles,repo_desc,contributor_count,language = [],[],[],[]

try:
    element = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[4]/main/div[3]/div/div[2]/article")))
    boxes = driver.find_elements(By.XPATH,"/html/body/div[4]/main/div[3]/div/div[2]/article")
    
    for box in boxes:
        try:
            val1 = box.find_element(By.TAG_NAME,"h1").text
        except NoSuchElementException:
            val1 = ""
        
        try:
            val2 = box.find_element(By.TAG_NAME,"p").text
        except NoSuchElementException:
            val2 = ""
        
        try:
            val3 = len(box.find_elements(By.XPATH,'.//span[@class="d-inline-block mr-3"]/a'))
        except NoSuchElementException:
            val3 = 0
            
        try:
            val4 = box.find_element(By.XPATH,'.//span[@itemprop="programmingLanguage"]').text
        except NoSuchElementException:
            val4 = ""
    
        repo_titles.append(val1)
        repo_desc.append(val2)
        contributor_count.append(val3)
        language.append(val4)
    
    df = pd.DataFrame()
    df["Title"] = repo_titles
    df["Description"] = repo_desc
    df["Contributors Count"] = contributor_count
    df["Language Used"] = language
    
    display(df)
    
except TimeoutException:
    print("Element Not Found!!")
```

### Question 6

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://www.billboard.com/")
driver.find_element(By.XPATH,'//*[@id="main-wrapper"]/header/div/div[2]/div/div/div[2]/div[2]/div/div/nav/ul/li[1]/a').click()
driver.find_element(By.XPATH,'//*[@id="main-wrapper"]/header/div/div[3]/div/nav/ul/li[1]/a').click()

titles , artists , last_week_rank , peak_rank , weeks_on_board = [] , [] , [] , [] , []

try:
    element = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME,"o-chart-results-list-row-container")))
    songs = driver.find_elements(By.CLASS_NAME,"o-chart-results-list-row-container")
    
    for song in songs:
        li = song.find_element(By.XPATH , './/li[@class="lrv-u-width-100p"]')
        titles.append(song.find_element(By.XPATH,'.//ul/li/h3').text)
        artists.append(li.find_element(By.XPATH,'.//ul/li/span').text)
        last_week_rank.append(li.find_element(By.XPATH,'.//ul/li[4]').text)
        peak_rank.append(li.find_element(By.XPATH,'.//ul/li[5]').text)
        weeks_on_board.append(li.find_element(By.XPATH,'.//ul/li[6]').text)
        
        
    df = pd.DataFrame()
    df["Song Title"] = titles
    df["Artists"] = artists
    df["Last Week Rank"] = last_week_rank
    df["Peak Rank"] = peak_rank
    df["Weeks on Board"] = weeks_on_board
    
    display(df)
    
    
except TimeoutException:
    print("Element Not Found!!")
    
```

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://www.naukri.com/")

# We are putting text inside the earch text box
input_ele = driver.find_element(By.CLASS_NAME,"suggestor-input ")
input_ele.click()
input_ele.send_keys("data science")
# Clicking the serach button
bttn = driver.find_element(By.CLASS_NAME,"qsbSubmit")
bttn.click()

# Getting job information cards
try:
    element = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'//article[@class="jobTuple bgWhite br4 mb-8"]')))
    cards = driver.find_elements(By.XPATH,'//article[@class="jobTuple bgWhite br4 mb-8"]')
    
    designations , company , location , skillsets = [] , [] , [] , []
    
    for card in cards:
        designations.append(card.find_element(By.XPATH,"./div[1]/div[1]/a").text)
        company.append(card.find_element(By.XPATH,"./div[1]/div[1]/div/a").text)
        location.append(card.find_element(By.XPATH,"./div[1]/div[1]/ul/li[3]/span").text)
        
        skills_ele = card.find_elements(By.XPATH,"./ul/li")
        skills = [skill.text for skill in skills_ele]
        skillsets.append(" | ".join(skills))
   
except TimeoutException:
    print("Eement Not Found")
```

```python
df = pd.DataFrame()
df["Designation"] = designations
df["Company"] = company
df["Location"] = location
df["Skills"] = skillsets

display(df)
```

### Question 8

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare")


rows = driver.find_elements(By.XPATH,'/html/body/div/div[2]/div[2]/div/div[2]/div/table/tbody/tr')

rank , title , author , sales , publisher , genre = [] , [] , [] , [] , [] , []

for row in rows:
    tds = row.find_elements(By.XPATH,'.//td')
    rank.append(tds[0].text)
    title.append(tds[1].text)
    author.append(tds[2].text)
    sales.append(tds[3].text)
    publisher.append(tds[4].text)
    genre.append(tds[5].text)
    
df = pd.DataFrame()
df["Rank"] = rank
df["Title"] = title
df["Author"] = author
df["Sales"] = sales
df["Publisher"] = publisher
df["Genre"] = genre

display(df)
```

### Question 9

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://www.imdb.com/list/ls095964455/")

items = driver.find_elements(By.XPATH,'//div[@class="lister-item mode-detail"]')

name , year_span , genre , run_time , ratings , votes = [] , [] , [] , [] , [] , [] 

for item in items:
    name.append(item.find_element(By.XPATH,'.//div[@class="lister-item-content"]/h3/a').text)
    year_span.append(item.find_element(By.XPATH,'.//div[@class="lister-item-content"]/h3/span[2]').text)
    genre.append(item.find_element(By.XPATH,'.//div[@class="lister-item-content"]/p/span[5]').text)
    run_time.append(item.find_element(By.XPATH,'.//div[@class="lister-item-content"]/p/span[3]').text)
    ratings.append(item.find_element(By.XPATH,'.//div[@class="lister-item-content"]/div[1]/div/span[2]').text)
    votes.append(item.find_element(By.XPATH,'.//div[@class="lister-item-content"]/p[4]/span[2]').text)
    
df = pd.DataFrame()
df["Name"] = name
df["Year Span"] = year_span
df["Genre"] = genre
df["Run Time"] = run_time
df["Ratings"] = ratings
df["Votes"] = votes

display(df)
```

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get("https://archive.ics.uci.edu/ml/index.php")
driver.find_element(By.XPATH,"/html/body/table[1]/tbody/tr/td[2]/span[2]/a").click()

rows = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr')

name , data_types , default_task , attribute_types , instances , attributes , year = [] , [] , [] , [] , [] , [] , []

for row in rows:
    tds = row.find_elements(By.XPATH,'./td')
    name.append(tds[0].text)
    data_types.append(tds[1].text)
    default_task.append(tds[2].text)
    attribute_types.append(tds[3].text)
    instances.append(tds[4].text)
    attributes.append(tds[5].text)
    year.append(tds[5].text)
    
df = pd.DataFrame()
df["Name"] = name
df["Data Types"] = data_types
df["Default Task"] = default_task
df["Attribute Types"] = attribute_types
df["Instances"] = instances
df["Attributes"] = attributes
df["Year"] = year
df = df.drop([0])

display(df)
```
